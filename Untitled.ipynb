## Leap year identification

```python
years = [x for x in range(1990, 2200)]
```

```python
for year in years:
    if year%4==0:
        if (year%100==0 & year%400==0):
            print(year, ": Leap Year")
        elif year%100!=0:
            print(year, ": leap year")
        else:
            print(year, ": Not leap year")
    else:
        print(year, ": Not leap year")
```
